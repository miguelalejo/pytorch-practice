Backpropagation using PyTorch
image.png

```python
tensor_x = torch.tensor(4.,requires_grad=True)
tensor_y = torch.tensor(-3.,requires_grad=True)
tensor_z = torch.tensor(5.,requires_grad=True)
```

```python
tensor_q = tensor_x + tensor_y
tensor_f = tensor_q * tensor_z
```

```python
tensor_f.backward()
```

```python
print(tensor_x.grad)
print(tensor_y.grad)
print(tensor_z.grad)
```
